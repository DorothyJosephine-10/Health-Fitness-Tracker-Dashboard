# Capping

```python
# capping for Age
q1=df["Age"].quantile(0.25)
print("Q1:",q1)
q3=df["Age"].quantile(0.75)
print("Q3:",q3)
iqr = q3-q1
print(iqr)
upper_fence = q3+1.5 * iqr
lower_fence = q1-1.5 * iqr
print("upper fence:",upper_fence)
print("lower fence:",lower_fence)
df["Age_capped"]=df["Age"]
df.loc[df["Age_capped"]>upper_fence,"Age_capped"]=upper_fence
df.loc[df["Age_capped"]<lower_fence,"Age_capped"]=lower_fence
```

```python
continuous_cols = ['Age', 'Weight (kg)', 'Height (m)', 'Max_BPM', 'Avg_BPM', 
                   'Resting_BPM', 'Session_Duration (hours)', 'Calories_Burned', 
                   'Fat_Percentage', 'Water_Intake (liters)', 'BMI']
```

```python
for col in continuous_cols:
    Q1 = final_df[col].quantile(0.25)
    Q3 = final_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - 1.5*IQR
    upper_fence = Q3 + 1.5*IQR
    outliers = final_df[(final_df[col] < lower_fence) | (final_df[col] > upper_fence)]
    print(f"{col}: {len(outliers)} outliers")
```

```python
def iqr_cap(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR
    # Cap values outside the fences
    return series.clip(lower=lower_fence, upper=upper_fence)

# Apply to all continuous columns
for col in continuous_cols:
    final_df[col] = iqr_cap(final_df[col])
```

```python
for col in continuous_cols:
    Q1 = final_df[col].quantile(0.25)
    Q3 = final_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - 1.5*IQR
    upper_fence = Q3 + 1.5*IQR
    outliers = final_df[(final_df[col] < lower_fence) | (final_df[col] > upper_fence)]
    print(f"{col}: {len(outliers)} outliers")
```

# Scatter plot for BMI and Fat_percentage by various workout

```python
plt.figure(figsize=(12, 6))
custom_colors = {
    "Strength": "#E74C3C",  # red
    "Cardio": "#3498DB",    # blue
    "HIIT": "#2ECC71",      # green
    "Yoga": "#9B59B6"       # purple
}

sns.scatterplot(
    x="BMI",
    y="Fat_Percentage",
    hue="Workout_Type",
    size="Workout_Frequency (days/week)",
    data=df1,
    palette=custom_colors,
    alpha=0.7,
    sizes=(40, 200)
)
plt.title("BMI vs Fat Percentage by Workout Type", fontsize=14)
plt.xlabel("BMI")
plt.ylabel("Fat Percentage")
plt.legend(title="Workout Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
